```python
#import modules
import promptMonitor
import imp
imp.reload(promptMonitor)
import matplotlib.pyplot as plt
```

```python
#The sliding-window method
#read the data from root files
prompt_far = '1permonth'
if prompt_far=='1permonth':
    nthr = 3
elif prompt_far=='1peryear':
    nthr = 4
thisMonitor_sliding = promptMonitor.promptMonitor()
thisMonitor_sliding.setSlidingPar(4, '1e-3', nthr)
thisMonitor_sliding.readPromptResult('sliding-window', 'intp1311.data', 0)
thisMonitor_sliding.readPromptResult('sliding-window', 'intp1311.data', 1)
thisMonitor_sliding.readPromptResult('sliding-window', 'intp3003.data', 0)
thisMonitor_sliding.readPromptResult('sliding-window', 'intp3003.data', 1)
thisMonitor_sliding.calEff()
thisMonitor_sliding.calTime()
```

```python
#draw the plots
thisDraw = promptMonitor.promptMonitor_draw()
thisDraw.setMonitor([thisMonitor_sliding])
thisDraw.drawEffciency('prompt_eff_sliding_%s'%prompt_far)
thisDraw.drawTriggerTime('prompt_time_sliding_%s'%prompt_far)
thisDraw.saveFigs()
plt.show()
```

```python
#the bayesian-block method
#read the data from root files
if prompt_far=='1permonth':
    ncp = 9.3
elif prompt_far=='1peryear':
    ncp = 16.2

thisMonitor_bayesian = promptMonitor.promptMonitor()
thisMonitor_bayesian.setBayesianPar(ncp)
thisMonitor_bayesian.readPromptResult('bayesian-block', 'intp1311.data', 0)
thisMonitor_bayesian.readPromptResult('bayesian-block', 'intp1311.data', 1)
thisMonitor_bayesian.readPromptResult('bayesian-block', 'intp3003.data', 0)
thisMonitor_bayesian.readPromptResult('bayesian-block', 'intp3003.data', 1)
thisMonitor_bayesian.calEff()
thisMonitor_bayesian.calTime()
```

```python
#draw the plots
thisDraw = promptMonitor.promptMonitor_draw()
thisDraw.setMonitor([thisMonitor_bayesian])
thisDraw.drawEffciency('prompt_eff_bayesian_%s'%prompt_far)
thisDraw.drawTriggerTime('prompt_time_bayesian_%s'%prompt_far)
thisDraw.saveFigs()
plt.show()
```

```python
#Draw the comparison between sliding and bayesian-block
thisDraw = promptMonitor.promptMonitor_draw()
thisDraw.setMonitor([thisMonitor_sliding, thisMonitor_bayesian])
thisDraw.drawEff_cmp('prompt_eff_cmp_%s'%prompt_far)
thisDraw.drawTime_cmp('prompt_time_cmp_%s'%prompt_far)
thisDraw.saveFigs()
```
